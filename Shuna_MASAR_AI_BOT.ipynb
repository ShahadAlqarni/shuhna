```python
#---------------------------->convert any html related to be understanded by telegram<---------------------------------
import re
def convert_markdown_to_html(text):

    text = re.sub(r'\*\*(.*?)\*\*', r'<b>\1</b>', text) #replace ** to BOLD
    text = re.sub(r'/ \\t',r"     ",text) #replace \t to "        " -spaces-
    return text

def send_welcome_message(chat_id,first_name):
    text=f"""مرحباً أ.{first_name} في تطبيق شُحنَة!  
يسرني أن اساعدك في نقل بضائعك بأمان وفعالية.  ماذا تريد أن تستفسر عنه اليوم؟  هل ترغب في:

* حساب تكلفة شحنة؟  (أخبرني بنوع الشحنة، وزنها أو حجمها، نقطة الانطلاق ونقطة الوصول) 💰
* تتبع شحنة موجودة؟ (أخبرني برقم الحجز الخاص بك) 🗺️
* معرفة المزيد عن خدماتنا؟ (أنواع الشحنات، التسعير، إلغاء الحجز، العروض...) 🚛
* الإبلاغ عن مشكلة أو شكوى؟ (سأحتاج إلى رقم الحجز الخاص بك) 🚨


سواء كنت تحتاج لنقل ثقيل أو خفيف، لدينا الحل المناسب لك!  تطبيقنا خريطة تتبع مباشرة عبر جوجل ماب لتتبع شحنتك في كل لحظة."""
    url_choices=main_buttons()
    bot.send_message(chat_id, text,reply_markup=url_choices)
```

```python
#----------------------------------> Prompt template <-----------------------------------
prompt_template = """
**مهمتك:** أنت مساعد ذكي لتطبيق "شُحنَة" المتخصص في نقل البضائع عبر السطحات.  
**تفاصيل المشروع (يجب الالتزام بها):**  
1. **أنواع الخدمات:**  
   - نقل ثقيل (أكثر من 5 أطنان).  
   - نقل خفيف (أقل من 5 أطنان).  
   - تتبع الشحنات عبر خرائط جوجل بشكل مباشر.  
2. **التسعير:**  
   - يعتمد على:  
     - المسافة (كل 1 كم = 2 ريال للنقل الخفيف، 5 ريال للنقل الثقيل).  
     - الحمولة (طن أو م³).  
   - مثال: نقل 3 أطنان من الرياض إلى جدة (950 كم) = 950 * 5 = 4,750 ريال.  
3. **سياسات الإلغاء والتعويض:**  
   - إلغاء مجاني قبل 24 ساعة من الموعد.  
   - تعويض 20% من التكلفة عند التأخير أكثر من 6 ساعات.  
4. **نظام التقييم:**  
   - العملاء يُقيّمون السائقين من 1 إلى 5 نجوم بناءً على:  
     - الالتزام بالوقت.  
     - حالة البضاعة عند التسليم.  
5. **العروض الموسمية:**  
   - خصم 30% على النقل داخل المدينة خلال فصل الصيف.  
   - نقل مجاني للحمولات الصغيرة (أقل من 1 طن) لأول طلب.  

**إرشادات الرد:**  
- استخدم لغة ودية مع إضافة رموز تعبيرية (🚛، ⏰، 💰).   
- في حالة الشكوى، اطلب **رقم الحجز** أولًا.  
- تجنب الردود العامة مثل "اتصل بالدعم".  

**أمثلة توضيحية (Two-Shot Learning):**  

**المثال 1 (استفسار عن التكلفة):**  
- المستخدم: "كم تكلفة نقل 4 أطنان من الدمام إلى أبو ظبي؟"  
- الرد:  
"مرحبًا! تكلفة النقل الثقيل (4 أطنان) تُحسب كالتالي:  
🚛 **المسافة**: 1,200 كم (من الدمام إلى أبو ظبي)  
💰 **التكلفة**: 1,200 كم * 5 ريال/كم = 6,000 ريال.  
"  

**المثال 2 (بلاغ عن تأخير):**  
- المستخدم: "الشحنة كانت من المفترض أن تصل الساعة 3 PM، والآن الساعة 8 PM ولا يوجد أي تحديث!"  
- الرد:  
"نعتذر بشدة عن هذا التأخير غير المقبول! 🚨  
لتصعيد الأمر، الرجاء إرسال:  
1. رقم الحجز (مثال: #1234).  
2. صورة من إشعار الشحن (إن وجد).  
3. وصف مفصل للمشكلة.  
سيتم تعويضك بنسبة 20% من التكلفة وفقًا لسياستنا، وسنتواصل معك خلال 30 دقيقة!"  

**المثال 3 (استفسار عن التتبع):**  
- المستخدم: "أين شحنتي الآن؟ رقم الحجز #5678."  
- الرد:  
"تم العثور على حجزك! 🗺️  
📍 **الموقع الحالي**: على طريق الرياض-جدة (قرب مدينة سدير).  
⏰ **الوقت المتوقع للوصول**: 7:00 PM.  
"  

**المثال 4 (استخدام عرض موسمي):**  
- المستخدم: "سمعت عن خصم 30%، كيف أستخدمه؟"  
- الرد:  
"أهلاً! خصم 30% متاح الآن للنقل داخل المدينة حتى 30 أغسطس. 🎉  
للاستفادة منه:  
1. اختر "نقل داخل المدينة" عند الحجز.  
2. أدخل كود الخصم: **SUMMER30**.  
3. تمتع بالتخفيض! 🚚"  

**الاستفسار الحالي:**  
{}
"""
```

```python
import telebot
from telebot import types
import logging
import os
from dotenv import load_dotenv
from transformers import pipeline  
import google.generativeai as genai 
import re
# --------------------------------------------------------

load_dotenv()

# --------------------> API Keys <--------------------
API_TOKEN = os.getenv("API_TOKEN")
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")  
# -------------------------------------------------------------------

bot = telebot.TeleBot(API_TOKEN)

# --------------------> CamelBert sentimental analysis <--------------------
sentiment_analyzer = pipeline(
    "sentiment-analysis",
    model="CAMeL-Lab/bert-base-arabic-camelbert-da-sentiment"
)

# --------------------> Gemini Generation model <--------------------
genai.configure(api_key=GEMINI_API_KEY)  
gemini_model = genai.GenerativeModel('gemini-1.5-flash')

#-----------------------------------------><--------------------------------------------
#-------------------------------->callback_handler<-------------------------------------
#-----------------------------------------><--------------------------------------------

@bot.callback_query_handler(func=lambda call: call.data in ["report_yes", "report_no"])
def handle_report_decision(call):
    if call.data == "report_yes":
        # إرسال رابط الفورم
        form_url = "https://forms.gle/FWgmAMKA9ptFtQwv5"
        bot.send_message(call.message.chat.id, f"الرجاء تعبئة النموذج: {form_url}")
    else:
        bot.send_message(call.message.chat.id, "حسنًا، هل مساعدة أخرى؟")

@bot.callback_query_handler(func=lambda call: call.data == "general_inquiry")
def handle_general_inquiry_info(call):
    general_inquiry_description = "اذكر كل استفساراتك وسأجيب عليك بكل ما تحتاج معرفته👍"

   
    bot.answer_callback_query(call.id)
    bot.send_message(call.message.chat.id, general_inquiry_description, parse_mode="HTML") #handle the today word message


@bot.callback_query_handler(func=lambda call:call.data=="send_welcome_message")
def handle_welcome_info(call):
    chat_id = call.message.chat.id
    username = call.from_user.first_name
    send_welcome_message(chat_id,username)
def main_buttons():
    general_inquiry = types.InlineKeyboardButton(
        text="لدي استفسار عام",
        callback_data="general_inquiry"
    )
    report = types.InlineKeyboardButton(
        text="اريد رفع بلاغ او مشكلة",
        url="https://forms.gle/FWgmAMKA9ptFtQwv5"
    )
    keyboard = types.InlineKeyboardMarkup([[general_inquiry, report]])
    return keyboard
def create_yes_no_keyboard():
    keyboard = types.InlineKeyboardMarkup()
    yes_btn = types.InlineKeyboardButton("نعم ✅", callback_data="report_yes")
    no_btn = types.InlineKeyboardButton("لا ❌", callback_data="report_no")
    keyboard.add(yes_btn, no_btn)
    return keyboard

@bot.message_handler(commands=['start'])
def send_welcome(message):
    username = message.from_user.first_name
    global send_start_command
    send_start_command = False
    send_welcome_message(message.chat.id, username)

@bot.message_handler(func=lambda message: True)
def handle_message(message):
    global send_start_command
    if not send_start_command:
        user_text = message.text

        
        # --------------------> 1) first step: Sentimental Analysis step <--------------------
        #Check if the user looks unsatisfied with the product and wants to send a report
        try:
            sentiment = sentiment_analyzer(user_text)[0]
            if sentiment['label'] == 'negative':
             
                bot.reply_to(message, 
                    "هل تواجه مشكلة تحتاج إلى رفع بلاغ؟",
                    reply_markup=create_yes_no_keyboard()
                )
                return
        except Exception as e:
            logging.error(f"Sentiment error: {e}")
        
        # --------------------> 2) second step: Generation step--------------------
        try:
          response = gemini_model.generate_content(
          prompt_template.format(user_text)  #gives the model the prompt
        )
          response = convert_markdown_to_html(response.text)
          bot.reply_to(message, response,parse_mode="HTML")
        except Exception as e:
            logging.error(f"Gemini error: {e}")
            bot.reply_to(message, "عذرًا، حدث خطأ في المعالجة.")
    else:
        send_welcome(message)
        send_start_command = False

if __name__ == "__main__":
    print("bot start working")
    try:
        send_start_command=True
        bot.infinity_polling() #to make it work all the time
    except Exception as e:
        logging.error(f"Polling error: {e}")
```
